Calculate $\tau_{drift}$ and $\tau_{diff}$ to determine how close these time constants are to the measured relaxation frequency$\omega_{D}$. This will help confirm whether the state is a trap peak or a dielectric relaxation peak.

```python
# sample length in cm
L = 40e-6

# amplitude of AC signal (V_{RMS}) in volts
V = 20e-3

# test temperatures
Tt = 300
T = np.linspace(200,300,num=1e3)

# constants
kB = 8.61733e-5
q = 1.602e-19

# test mobilities
muT = 1e-4
mu = np.linspace(1e-5,1e-2,num=1e3)

# calculate the time constants assuming a range of mobilities and T = 300K (calculation 1)
tau_drift1 = (L**2)/(mu*V)
tau_diff1 = (L**2)/(mu*kB*Tt)

omega_drift1 = (2*np.pi)/tau_drift1
omega_diff1 = (2*np.pi)/tau_diff1

# calculate the time constants assuming a range of temperatures and mobility = 1e-4 (calculation 2)
tau_drift2 = (L**2)/(muT*V)
tau_diff2 = (L**2)/(muT*kB*T)

omega_drift2 = (2*np.pi)/tau_drift2
omega_diff2 = (2*np.pi)/tau_diff2

# plot calculation 1
fig,ax = plt.subplots(1,1)
ax.plot(mu,omega_drift1,'o', markersize = 10) 
ax.plot(mu,omega_diff1,'o', markersize = 10) 
ax.set_xlabel("$\\mu$ ($cm^{2}V^{-1}s^{-1}$)")
ax.set_ylabel("$\\omega$ ($rad/s$)")
ax.set_xscale("log") 
ax.set_yscale("log") 
# ax.set_ylim([1e-3,3e-1]);


# plot calculation 2
fig,ax = plt.subplots(1,1)
# ax.plot(T,omega_drift2,'o', markersize = 10) 
ax.plot(T,omega_diff2,'o', markersize = 10) 
ax.set_xlabel("$T$ ($K$)")
ax.set_ylabel("$\\omega$ ($rad/s$)")
# ax.set_xscale("log") 
ax.set_yscale("log") 
# ax.set_ylim([1e-3,3e-1]);

print(omega_drift2)
```

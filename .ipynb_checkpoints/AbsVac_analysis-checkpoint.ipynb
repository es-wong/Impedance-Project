```python
os.chdir(r'C:\Users\Eric\Desktop\Lab_Work\admittance\ligand_and_interface\analysis_work\EDT_schottky\09_2017_m1\data\opticalSupplement\optical_vacData')

os.listdir()
nPoints = 1201
nCols = 2
nData = 360

master = np.zeros((nData,nPoints,3))




# the tmp data has the following format:
#      index 0 is wavelength in nm
#      index 1 is absorption in arbitrary units
#      index 2 is normalized absorption

# import as made data into master array
file1 = 'vacEvolution2.csv'     
tmp1 = np.genfromtxt(file1,delimiter = ',')

for sample in range(0,nData): 
    master[sample,:,0] = tmp1[2:,(2*sample)] #import wavelength
    master[sample,:,1] = tmp1[2:,(2*sample)+1] #import absorbance
            

# normalize the as made data by finding an index that cuts out the higher excitonic features and normalizng to max of this range
absIdx = np.where(master[0,:,0] <=920)[0][0]
MaxArray = np.zeros(nData)

for sample in range(0,nData):
    MaxArray[sample] = max(master[sample,:absIdx,1])
    master[sample,:,2] = master[sample,:,1]/MaxArray[sample]

    
def sampleAbs(k):
    return master[k,:,2]

def waveLength(dummy):
    return master[0,:,0]

# indicate the figures directory
figDir = r'C:\Users\Eric\Desktop\Lab_Work\admittance\ligand_and_interface\analysis_work\EDT_schottky\09_2017_m1\data\opticalSupplement\optical_vacData\figs'
```

```python
# plot the data to see what we are dealing with

fig,ax = plt.subplots(1,1)
for sample in range(0,nData):
    ax.plot(waveLength(1),sampleAbs(sample),'o', markersize = 15)
ax.set_xlabel("$\\lambda$ ($nm$)")
ax.set_ylabel("$Absorption$ ($arb$) ")
# ax.set_xscale("log") 
# ax.set_yscale("log")
ax.set_xlim([800,1200])
ax.set_ylim([0,1.5])
ax.xaxis.set_major_locator(MaxNLocator(4))
```

```python
import lmfit as lmf
from lmfit.models import ExponentialModel, GaussianModel

x = waveLength(1)[:absIdx]

yInit = sampleAbs(0)[:absIdx] 

exp_mod = ExponentialModel(prefix='exp_')
pars1 = exp_mod.guess(yInit, x=x)

gauss1 = GaussianModel(prefix='g1_')
pars1.update(gauss1.make_params())

pars1['g1_center'].set(980, min=950, max=1100)
pars1['g1_sigma'].set(30, min=3)
pars1['g1_amplitude'].set(1, min=0.1)

mod = gauss1 + exp_mod

# empty arrays for storing fit parameters
centerFits = []
sigmaFits = []
ampFits = []

fits = []

init1 = mod.eval(pars1, x=x)
fits.append(mod.fit(yInit, pars1, x=x))

# run the fit
centerFits.append(fits[0].params['g1_center'].value)
sigmaFits.append(fits[0].params['g1_sigma'].value)
ampFits.append(fits[0].params['g1_amplitude'].value)


# now run fit in a loop
for sample in range(1,nData):
    fits.append(mod.fit(sampleAbs(sample)[:absIdx],pars1,x=x))
    
    centerFits.append(fits[sample-1].params['g1_center'].value)
    sigmaFits.append(fits[sample-1].params['g1_sigma'].value)
    ampFits.append(fits[sample-1].params['g1_amplitude'].value)
    
    
#     params = lmf.Parameters() 
#     params.add('Rs',value = RsFits[t-1], min = 0, max = 5)
#     params.add('Rct',value = RctFits[t-1], min = 0, max = 100)
#     params.add('Q',value = QFits[t-1], min = 0, max = 1e-3)
#     params.add('alpha',value = alphaFits[t-1] , min = 0, max = 1)

#     # run the fit
#     fit = lmf.minimize(circuitResidQ,params,args = (w(t+offSet)[bounds[t]:],Zr(t+offSet)[bounds[t]:],Zi(t+offSet)[bounds[t]:]))
#     RsFits.append(fit.params['Rs'].value)
#     RctFits.append(fit.params['Rct'].value)
#     QFits.append(fit.params['Q'].value)
#     alphaFits.append(fit.params['alpha'].value)
#     fit.params
```

```python
# Make a gaussian DOS out of the fit parameters. First convert the center to eV. 

eVcenter = 1240/center

sigmaDist = center + sigma
sigmaDist_eV = 1240/sigmaDist

eVsigma = abs(sigmaDist_eV - eVcenter)

# now make Gaussian DOS of this function in energy
gaussDOS = np.zeros((nData,100))

# create energy scale
E = np.linspace(1,1.4,100)

# perform calculation
for sample in range(0,nData):
    gaussDOS[sample,:] = np.exp(-(E-eVcenter[sample])**2/(2*(eVsigma[sample]**2)))

```

One last thing: let's think about the amount of charge *in* the band by thinking about what's going on with $E_{F}$. Assume $E_{F}$ is pinned $\sim 0.2$ eV above the band edge. How does the red-shift compete wit the shrinking of the width? 

If the decrease in the width of $E$ wins out, then our qualitative analysis at least checks out. If not, then we'll have to think harder. 

```python
# input parameters necessary for fermi function 
Ef1 = np.zeros(nData)
kB = 8.6173e-5
Tt = 300
Efstar = 1.0

# like in the analysis done for the paper, let's calculate overlap

def fermi(energy,Ef,temp):
    return 1/(1 + np.exp((energy-Ef)/(kB*temp)))

def fProd(f1, f2):
    return f1*f2

def fIntegrand(energy,center,sigmaT,Ef,temp):
    return np.exp(-(energy-center)**2/(2*(sigmaT**2)))*(1/(1 + np.exp((energy-Ef)/(kB*temp))))



# gaussDOS[sample,:] = np.exp(-(E-eVcenter[sample])**2/(2*(eVsigma[sample]**2)))

# def fIntegrand(energy,a,var,Ef,temp):
#     return (a*np.exp(-(energy)**2/(2*var**2)))*(1/(1 + np.exp((energy-Ef)/(kB*temp))))

# 
#     fermiInt = quad(fIntegrand, E_star, np.inf, args = (amp,sigmaVar,E_star,300))


f2 = fermi(E,Ef,Tt)

fermiOverlap = np.zeros((nData,100))
fermiInt1 = np.zeros(nData)
fermiInt2 = np.zeros(nData)

fermiFunc = np.zeros((nData,100))

for sample in range(0,nData):
    Ef1[sample] = eVcenter[sample] - Efstar
    fermiFunc[sample,:] = fermi(E,Ef1[sample],Tt)
    
    fermiOverlap[sample,:] = fProd(gaussDOS[sample,:],fermiFunc[sample,:])

    fermiInt1[sample] = np.trapz(fermiOverlap[sample,:])
    
    fermiInt2[sample] = quad(fIntegrand,Ef1[sample],np.inf, args = (eVcenter[sample],eVsigma[sample],Ef1[sample],Tt))[0]



baseline1 = max(fermiInt1)
baseline2 = max(fermiInt2)

# plot integral of overlap vs. time
fig,ax = plt.subplots(1,1)
for sample in range(0,nData):
    ax.plot(time[sample],fermiInt1[sample]/baseline1,'o', markersize = 15,color = tuple(colSet[sample,:]))
ax.set_xlabel("$t$ ($min$)")
ax.set_ylabel("$\\Delta n_{V}$ ($\%$) ")
# ax.set_xscale("log") 
# ax.set_yscale("log")
# ax.set_xlim([800,1200])
# ax.set_ylim([0,1.5])
ax.xaxis.set_major_locator(MaxNLocator(4))
ax.yaxis.set_major_locator(MaxNLocator(4))
# figName = '\\FWHM_evolution.png'
# plt.savefig(figDir + figName,dpi = 300,bbox_inches = 'tight')

# plot integral of overlap vs. time
fig,ax = plt.subplots(1,1)
for sample in range(0,nData):
    ax.plot(time[sample],fermiInt2[sample]/baseline2,'o', markersize = 15,color = tuple(colSet[sample,:]))
ax.set_xlabel("$t$ ($min$)")
ax.set_ylabel("$\\Delta n_{V}$ ($\%$) ")
# ax.set_xscale("log") 
# ax.set_yscale("log")
# ax.set_xlim([800,1200])
ax.set_ylim([0.7,1.1])
ax.xaxis.set_major_locator(MaxNLocator(4))
ax.yaxis.set_major_locator(MaxNLocator(4))
figName = '\\nv_evolution.png'
plt.savefig(figDir + figName,dpi = 300,bbox_inches = 'tight')

```
